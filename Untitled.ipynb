```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("dataset/dataset.csv")
```

```python
tanggal = dataset['tanggal']
x = dataset['index']
y_pertambahan = dataset['jumlah pertambahan']
y_total = dataset['jumlah total']
```

### Plot data berdasarkan pertambahan perhari

```python
plt.plot(x,y_pertambahan)
plt.xlabel('hari ke') 
plt.ylabel('Pertambahan perhari') 
plt.show()
```

### Plot data berdasarkan total semua kasus perhari

```python
plt.plot(x,y_total)
plt.xlabel('hari ke') 
plt.ylabel('Tota Kasus') 
plt.show()
```

```python
dates = (
    '17 April 2020',
    '18 April 2020',
    '19 April 2020',
    '20 April 2020',
    '21 April 2020',
)
```

### Linear Regression Introduction

![title](img/rumus.PNG)

```python
def estimate_coef(x, y): 
    n = np.size(x) 
  
    b_1 = (n * np.sum(x*y) - np.sum(x) * np.sum(y)) / (n * np.sum(x*x) - np.power(np.sum(x),2))
    b_0 = (np.sum(y) - b_1 * np.sum(x)) / n
  
    return [b_0, b_1]

def plot_regression_line(x, y, b, x_label='', y_label=''): 
    plt.scatter(x, y, color = "m", marker = "o", s = 30) 

    y_pred = b[0] + b[1]*x 
  
    plt.plot(x, y_pred, color = "g") 
  
    plt.xlabel(x_label) 
    plt.ylabel(y_label) 
  
    plt.show()
    
def predict(b, title=''):
    print(title)
    x_ = 46
    for tanggal in dates:
        y_ = b[0] + b[1]*x_
        print(tanggal ,':',int(y_),' kasus')
        x_ += 1
        
def calculate_r_square(x, y_obs, b, title=''):
    y_pred = []
    for _x in x:
        _y = b[0] + b[1] * _x
        y_pred.append(_y)
    
    y_pred = np.array(y_pred)
    y_mean = np.mean(y_obs)
    sse = np.sum(np.power(y_obs - y_pred,2))
    tss = np.sum(np.power(y_obs - y_mean,2))
    r_square = 1 - (sse / tss)
    print('Nilai R Square',title, 'adalah :', r_square)
```

### Grafik berdasarkan jumlah total perhari

```python
b_total = estimate_coef(x, y_total)
print('Grafik berdasarkan jumlah total perhari')
print("Parameter :\nb_0 = {}  \nb_1 = {}".format(b_total[0], b_total[1])) 
plot_regression_line(x, y_total, b_total, 'hari ke-', 'Total Kasus')
```

### Prediksi Total Kasus covid-19 lima hari selanjutnya setelah tanggal 16 April 2020

```python
predict(b_total, 'Total Kasus Pada Tanggal ->')
```

### Grafik berdasarkan pertambahan kasus perhari

```python
b_pertambahan = estimate_coef(x, y_pertambahan)
print('Grafik berdasarkan pertambahan kasus perhari')
print("Parameter :\nb_0 = {}  \nb_1 = {}".format(b_pertambahan[0], b_pertambahan[1])) 
plot_regression_line(x, y_pertambahan, b_pertambahan, 'Hari ke-', 'Pertambahan Kasus Perhari')
```

### Prediksi Pertambahan Kasus covid-19 perhari dalam lima hari selanjutnya setelah tanggal 16 April 2020

```python
predict(b_pertambahan, 'Pertambahan Kasus Baru Pada Tanggal ->')
```

### Perhitungan R Square model Total Kasus Perhari

```python
calculate_r_square(x, y_total, b_total, 'Total Kasus Perhari')
```

### Perhitungan R Square model Pertambahan Kasus Perhari

```python
calculate_r_square(x, y_pertambahan, b_pertambahan, 'Pertambahan Kasus Perhari')
```
